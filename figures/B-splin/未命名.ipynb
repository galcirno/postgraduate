```python
x = np.linspace(-1,3,120)
interval0 = [1 if (i<1 and i>=0) else 0 for i in x]
interval1 = [1 if (i>=1 and i<2) else 0 for i in x]
interval2 = [1 if (i>=2 and i<0) else 0 for i in x]
y = x* interval0 +(2-x)* interval1 + x*0*interval2

plt.plot(x,y,label="$B_{2}(k)$")
# plt.show()
plt.xlabel("x")
plt.ylabel("y")

# change y range
plt.ylim(-0.25, 1.25)
plt.legend(loc='best')
plt.savefig("B2.pdf")
```

```python
x = np.linspace(-1,4,150)
interval0 = [1 if (i<1 and i>=0) else 0 for i in x]
interval1 = [1 if (i>=1 and i<2) else 0 for i in x]
interval2= [1 if (i>=2 and i<3) else 0 for i in x]
interval3 = [1 if (i>=3 and i<0) else 0 for i in x]
y = 1/2*x*x* interval0 +(1/2+(x-1)-(x-1)*(x-1))* interval1 + (1/2-(x-2)+1/2*(x-2)*(x-2))* interval2+x*0*interval3

plt.plot(x,y,label="$B_{3}(k)$")
# plt.show()
plt.xlabel("x")
plt.ylabel("y")

# change y range
plt.ylim(-0.2, 1.1)
plt.legend(loc='best')
plt.savefig("B3.pdf")
```
